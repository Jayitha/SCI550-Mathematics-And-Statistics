# Question 3

For the 7 samples of 5 housefly wing lengths & for milk yields, compute 95% confidence limits to the parametric mean for each sample and for the total sample based on 35 items. Base the standard errors of the means on the parametric standard deviations of these populations:  
Housefly wing lengths σ = 3.90  
Milk yields σ = 11.1597  
Record how many in each of the four classes of confidence limits (wing lengths & milk yields, n = 5 & n = 35) were correct – i.e., contained the parametric mean of the population?

```python
import numpy as np
import plotly.graph_objects as go
from scipy.stats import kurtosis, skew
from scipy.stats import anderson
import random
import statistics
```

```python
# Reading Input

jerseyCows = []
wingsHouseflies = []

with open("jersey cows", "r") as f:
    jerseyCows = list(map(int,f.read().split()));
with open("wings houseflies", "r") as f:
    wingsHouseflies = list(map(int,f.read().split()));

print(jerseyCows)
print(wingsHouseflies)
print(len(jerseyCows))
print(len(wingsHouseflies))
```

```python
# Sampling 7 samples of size 5 each from both

jerseyCowsSamples = []
wingsHousefliesSamples = []
k = 35
for _ in range(k):
    jerseyCowsSamples.append(jerseyCows[random.randint(0, len(jerseyCows)-1)])
    wingsHousefliesSamples.append(wingsHouseflies[random.randint(0,len(wingsHouseflies)-1)])
jerseyCowsSamples = np.asarray(jerseyCowsSamples).reshape((7,5))
wingsHousefliesSamples = np.asarray(wingsHousefliesSamples).reshape((7,5))
print("Samples obtained are: ")
print(jerseyCowsSamples)
print(wingsHousefliesSamples)
```

```python
# Computing 95% confidence interval for each sample

jerseyCowsSamplesIntervals = []
wingsHousefliesSamplesIntervals = []

sigma_wing = 3.90
sigma_milk = 11.1597

for i in range(7):
    mean_milk = np.mean(jerseyCowsSamples[i,:])
    mean_wing = np.mean(wingsHousefliesSamples[i,:])
    
    jerseyCowsSamplesIntervals.append(mean_milk - 1.95*sigma_milk/(5**0.5))
    jerseyCowsSamplesIntervals.append(mean_milk + 1.95*sigma_milk/(5**0.5))
    wingsHousefliesSamplesIntervals.append(mean_wing - 1.95*sigma_wing/(5**0.5))
    wingsHousefliesSamplesIntervals.append(mean_wing + 1.95*sigma_wing/(5**0.5))
    
jerseyCowsSamplesIntervals = np.asarray(jerseyCowsSamplesIntervals).reshape((7,2))
wingsHousefliesSamplesIntervals = np.asarray(wingsHousefliesSamplesIntervals).reshape((7,2))

print("95% Confidence intervals for each are: ")
print(wingsHousefliesSamplesIntervals)
print(jerseyCowsSamplesIntervals)
```

```python
# Finding number of samples within which mean occured

mean_milk = np.mean(jerseyCows)
mean_wing = np.mean(wingsHouseflies)

print("Milk Yeild Mean: ", mean_milk)
print("Wing Length Mean: ", mean_wing)

count1 = 0
count2 = 0
for i in range(7):
    if mean_milk > jerseyCowsSamplesIntervals[i,0] and mean_milk < jerseyCowsSamplesIntervals[i,1]:
        count1 = count1 + 1
    if mean_wing > wingsHousefliesSamplesIntervals[i,0] and mean_wing < wingsHousefliesSamplesIntervals[i,1]:
        count2 = count2 + 1

print("Milk Yeild: ",count1,"/7 samples contained population mean")
print("Wing Length: ",count2,"/7 samples contained population mean")
```

# Question 4

For 200 samples of n = 5 and n = 35 wing lengths, compute 95% and 99% confidence limits of the parametric mean employing the parametric standard error of the mean.  
Housefly wing lengths σ = 3.90  
Milk yields σ = 11.1597  
Out of 200 confidence intervals, how many contain the parametric mean of the population in the two cases for two sample sizes?  
Do you see any effect of increasing the sample size?  

```python
# Sampling 200 samples of size 5 and 35
wingsHousefliesSamples5 = []
wingsHousefliesSamples35 = []
for _ in range(200):
    for __ in range(5):
        wingsHousefliesSamples5.append(wingsHouseflies[random.randint(0,len(wingsHouseflies)-1)])
    for __ in range(35):
        wingsHousefliesSamples35.append(wingsHouseflies[random.randint(0,len(wingsHouseflies)-1)])
        
wingsHousefliesSamples5 = np.asarray(wingsHousefliesSamples5).reshape((200,5))
wingsHousefliesSamples35 = np.asarray(wingsHousefliesSamples35).reshape((200,35))
print("Samples obtained are: ")
print(wingsHousefliesSamples5)
print(wingsHousefliesSamples35)
```

```python
# Computing 95% & 99% confidence interval for each sample

wingsHousefliesSamples5Intervals95 = []
wingsHousefliesSamples35Intervals95 = []
wingsHousefliesSamples5Intervals99 = []
wingsHousefliesSamples35Intervals99 = []

sigma_wing = 3.90

for i in range(200):
    mean_5 = np.mean(wingsHousefliesSamples5[i,:])
    mean_35 = np.mean(wingsHousefliesSamples35[i,:])

    wingsHousefliesSamples5Intervals95.append(mean_5 - 1.95*sigma_wing/(5**0.5))
    wingsHousefliesSamples5Intervals95.append(mean_5 + 1.95*sigma_wing/(5**0.5))
    wingsHousefliesSamples5Intervals99.append(mean_5 - 2.58*sigma_wing/(5**0.5))
    wingsHousefliesSamples5Intervals99.append(mean_5 + 2.58*sigma_wing/(5**0.5))
    
    wingsHousefliesSamples35Intervals95.append(mean_35 - 1.95*sigma_wing/(35**0.5))
    wingsHousefliesSamples35Intervals95.append(mean_35 + 1.95*sigma_wing/(35**0.5))
    wingsHousefliesSamples35Intervals99.append(mean_35 - 2.58*sigma_wing/(35**0.5))
    wingsHousefliesSamples35Intervals99.append(mean_35 + 2.58*sigma_wing/(35**0.5))
    
wingsHousefliesSamples5Intervals95 = np.asarray(wingsHousefliesSamples5Intervals95).reshape((200,2))
wingsHousefliesSamples5Intervals99 = np.asarray(wingsHousefliesSamples5Intervals99).reshape((200,2))
wingsHousefliesSamples35Intervals95 = np.asarray(wingsHousefliesSamples35Intervals95).reshape((200,2))
wingsHousefliesSamples35Intervals99 = np.asarray(wingsHousefliesSamples35Intervals99).reshape((200,2))

print("95% Confidence intervals for Samples of size 5 are: ")
print(wingsHousefliesSamples5Intervals95)
print("99% Confidence intervals for Samples of size 5 are: ")
print(wingsHousefliesSamples5Intervals99)
print("95% Confidence intervals for Samples of size 35 are: ")
print(wingsHousefliesSamples35Intervals95)
print("99% Confidence intervals for Samples of size 35 are: ")
print(wingsHousefliesSamples5Intervals99)
```

```python
# Finding number of samples within which mean occured

mean_wing = np.mean(wingsHouseflies)
print("Wing Length Mean: ", mean_wing)

count595 = 0
count599 = 0
count3595 = 0
count3599 = 0

for i in range(200):
    if mean_wing > wingsHousefliesSamples5Intervals95[i,0] and mean_wing < wingsHousefliesSamples5Intervals95[i,1]:
        count595 = count595 + 1
    if mean_wing > wingsHousefliesSamples5Intervals99[i,0] and mean_wing < wingsHousefliesSamples5Intervals99[i,1]:
        count599 = count599 + 1
    if mean_wing > wingsHousefliesSamples35Intervals95[i,0] and mean_wing < wingsHousefliesSamples35Intervals95[i,1]:
        count3595 = count3595 + 1
    if mean_wing > wingsHousefliesSamples35Intervals99[i,0] and mean_wing < wingsHousefliesSamples35Intervals99[i,1]:
        count3599 = count3599 + 1  

print("N: 5 %: 95: ",count595,"/200 samples contained population mean")
print("N: 5 %: 99: ",count599,"/200 samples contained population mean")
print("N: 35 %: 95: ",count3595,"/200 samples contained population mean")
print("N: 35 %: 99: ",count3599,"/200 samples contained population mean")
```

# Comments

As the sample size increases, more samples "contain" the population mean.
