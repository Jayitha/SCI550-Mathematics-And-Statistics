# Question 1
Construct t-distribution and ÔÅ£2-distribution by sampling (with replacement) from the wing lengths of house flies for sample sizes: (a) 1400 samples of k = 5, and (b) 200 samples of k = 35. Compare these two distributions with the sampling distribution of means (computed earlier for the two sample sizes). Submit all the three plots.

```python
import numpy as np
import plotly.graph_objects as go
from scipy.stats import kurtosis, skew
from scipy.stats import anderson
import random
import statistics
```

```python
# Reading Input

wingsHouseflies = []

with open("wings houseflies", "r") as f:
    wingsHouseflies = list(map(int,f.read().split()));

wing_mean = np.mean(wingsHouseflies)
wing_stdev = statistics.stdev(wingsHouseflies)
print(wingsHouseflies)
print(len(wingsHouseflies))
```

```python
# Sampling 1400 samples with n = 5 and 200 samples with n = 35

wings5 = []
wings35 = []

for _ in range(1400):
    for __ in range(5):
        wings5.append(wingsHouseflies[random.randint(0,len(wingsHouseflies)-1)])
for _ in range(200):
    for __ in range(35):
        wings35.append(wingsHouseflies[random.randint(0,len(wingsHouseflies)-1)])

wings5 = np.asarray(wings5).reshape((1400,5))
wings35 = np.asarray(wings35).reshape((200,35))

print("K = 1400, n = 5: ")
print(wings5)
print("K = 200, n = 35: ")
print(wings35)
```

```python
# Constructing t distribution

t5 = [(np.mean(wings5[i,:]) - wing_mean)*(5**0.5)/(statistics.stdev(wings5[i,:]+0.000000001)) for i in range(1400)]
t35 = [(np.mean(wings35[i,:]) - wing_mean)*(35**0.5)/(statistics.stdev(wings35[i,:]+0.000000001)) for i in range(200)]

fig = go.Figure()
fig.add_trace(go.Histogram(
    x=t5,
    histnorm='probability',
    marker_color='#091b52',
    opacity=0.75
))

fig.update_layout(
    title_text='Distribution of t of Wing Lengths of Houseflies K = 1400 N = 5', # title of plot
    xaxis_title_text='t', # xaxis label
    yaxis_title_text='Probability', # yaxis label
)

fig.show()

fig = go.Figure()
fig.add_trace(go.Histogram(
    x=t35,
    histnorm='probability',
    marker_color='#091b52',
    opacity=0.75
))

fig.update_layout(
    title_text='Distribution of t of Wing Lengths of Houseflies K = 200 N = 35', # title of plot
    xaxis_title_text='t', # xaxis label
    yaxis_title_text='Probability', # yaxis label
)

fig.show()
```

```python
# Constructing chi2 distribution

chi25 = [5*statistics.stdev(wings5[i,:])**2/wing_stdev**2 for i in range(1400)]
chi235 = [35*statistics.stdev(wings35[i,:])**2/wing_stdev**2 for i in range(200)]

fig = go.Figure()
fig.add_trace(go.Histogram(
    x=chi25,
    histnorm='probability',
    marker_color='#091b52',
    opacity=0.75
))

fig.update_layout(
    title_text='Distribution of chi2 of Wing Lengths of Houseflies K = 1400 N = 5', # title of plot
    xaxis_title_text='chi2', # xaxis label
    yaxis_title_text='Probability', # yaxis label
)

fig.show()

fig = go.Figure()
fig.add_trace(go.Histogram(
    x=chi235,
    histnorm='probability',
    marker_color='#091b52',
    opacity=0.75
))

fig.update_layout(
    title_text='Distribution of chi2 of Wing Lengths of Houseflies K = 200 N = 35', # title of plot
    xaxis_title_text='chi2', # xaxis label
    yaxis_title_text='Probability', # yaxis label
)

fig.show()
```

Next we notice that the chi2 distribution isn't symmetric at all (although with more dof it gradually tends to normality). We also notice that unlike the z statistic and the t statistic, chi2 only takes on positive values.
